```python
import matplotlib.pyplot as plt

# Create sample dataset
intervals = ['1-17', '18-34', '35-51', '52-68', '69-85']
frequencies = [12, 25, 30, 20, 13]

plt.bar(intervals, frequencies, color='skyblue')
plt.xlabel("Age Interval")
plt.ylabel("Age Frequencies Distribution")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show
```

### KDE Plot(Known as Probability Density function)

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.kdeplot(data = df, x='bedrooms', fill=True)

plt.title('Housing price distribution')
plt.xlabel('Bedrooms')
plt.ylabel('Density')
plt.show()
```

### Histogram plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(data=df, x="price", kde=True, bins=30, color='skyblue')
plt.title("Housing price distribution with kde")
plt.xlabel("Price")
plt.ylabel("Frequency / Density")
```
