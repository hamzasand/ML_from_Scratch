### KDE Plot(Known as Probability Density function)

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.kdeplot(data = df, x='bedrooms', fill=True)

plt.title('Housing price distribution')
plt.xlabel('Bedrooms')
plt.ylabel('Density')
plt.show()
```

### Histogram plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(data=df, x="price", kde=True, bins=30, color='skyblue')
plt.title("Housing price distribution with kde")
plt.xlabel("Price")
plt.ylabel("Frequency / Density")
```

### Scatter plot
Scatter plot used to visualized the the relationshipe between two variables/atributes

```python
import seaborn as sns
import matplotlib.pyplot

sns.scatterplot(data=df, x="area", y="price")
plt.title("Price vs Area ")
plt.xlabel("Area")
plt.ylabel("Price")
```

### Table to view relationshipe between multiple atributes
